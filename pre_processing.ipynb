```python
import pickle
```

```python
with open('/Users/zehaodong/Desktop/NBA_RNN/jumpshot_possdata.pickle','rb') as f:
    poss_data = pickle.load(f)
```

```python
len(poss_data)
```

```python
poss_data[0]
```

```python
ball_handler = {}
for poss in poss_data:
    key = poss[0][5]
    if ball_handler.get(key) is None:
        ball_handler[key] = 1
    else:
        ball_handler[key] += 1
```

```python
len(ball_handler)
```

```python
sorted_ball_handler = sorted(ball_handler.items(), key = lambda value: value[1], reverse=True)
```

```python
sorted_ball_handler[:10]
```

```python
top10_ball_handler_list = [item[0] for i,item in enumerate(sorted_ball_handler) if i < 10]
```

```python
top10_ball_handler_list
```

```python
selected_poss_for_top10 = []
for poss in poss_data:
    if poss[0][5] in top10_ball_handler_list:
        selected_poss_for_top10.append(poss)
print(len(selected_poss_for_top10))
```

```python
poss_index_with_error = []
for i,poss in enumerate(selected_poss_for_top10):
    for moment in poss:
        def_player = moment[8]
        if len(def_player) == 0:
            poss_index_with_error.append(i)
            break
print(len(poss_index_with_error))
```

```python
poss_for_top10_after_correction = [poss for i,poss in enumerate(selected_poss_for_top10) if i not in poss_index_with_error]
print(len(poss_for_top10_after_correction))
```

```python
def_player_list_in_selected_poss = []
for i,poss in enumerate(poss_for_top10_after_correction):
    for moment in poss:
        def_player = moment[8]
        if len(def_player) == 0:
            print(moment)
        if def_player not in def_player_list_in_selected_poss:
            def_player_list_in_selected_poss.append(def_player)
print(len(def_player_list_in_selected_poss))
```

```python
with open('poss_data_for_top10.pickle', 'wb') as f:
    pickle.dump(poss_for_top10_after_correction,f)
```

```python
with open('top10_off_players.pickle','wb') as f:
    pickle.dump(sorted_ball_handler[:10], f)
```

```python
with open('all_def_players_in_selected_poss_for_top10.pickle', 'wb') as f:
    pickle.dump(def_player_list_in_selected_poss, f)
```
